```python
class Cronograma(ttk.Frame):
    def __init__(self, parent=None):
        ttk.Frame.__init__(self, parent, class_="Cronograma")
        #self.grid(sticky=N+E+S+W)
        #self.grid_remove()
        self.parent = parent
        self.CreateWidgets()
        
    def CreateWidgets(self):
        self.ListCols = ["Identidade","Nome","Ordem de Chegada"]
        self.ListTeste = [
            ("000","Teste0","Ausente"),
            ("001","Teste1","Ausente"),
            ("002","Teste2","1")
        ]
        
        self.Lista = ttk.Treeview(self, columns=self.ListCols, show="headings")
        #Configurar barras de rolagem
        vsb = ttk.Scrollbar(self, orient="vertical", command=self.Lista.yview)
        hsb = ttk.Scrollbar(self, orient="horizontal", command=self.Lista.xview)
        self.Lista.configure(yscrollcommand=vsb.set, xscrollcommand=hsb.set)
        
        self.Lista.grid(column=0, row=0, sticky='nsew')
        vsb.grid(column=1, row=0, sticky='ns')
        hsb.grid(column=0, row=1, sticky='ew')
        self.grid_columnconfigure(0, weight=1)
        self.grid_rowconfigure(0, weight=1)
        
        #preencher a lista
        for col in self.ListCols:
            self.Lista.heading(col, text=col.title(),
                command=lambda c=col: self.sortby(self.Lista, c, 0))
            # adjust the column's width to the header string
            self.Lista.column(col,
                width=font.Font().measure(col.title()))
        for item in self.ListTeste:
            self.Lista.insert('', 'end', values=item)
            # adjust column's width if necessary to fit each value
            #for ix, val in enumerate(item):
                #col_w = font.Font().measure(val)
                #if self.Lista.column(self.ListTeste[ix],width=None)<col_w:
                    #self.Lista.column(self.ListTeste[ix], width=col_w)
    
    def sortby(tree, col, descending):
        """sort tree contents when a column header is clicked on"""
        # grab values to sort
        data = [(tree.set(child, col), child) \
            for child in tree.get_children('')]
            # if the data to be sorted is numeric change to float
            #data =  change_numeric(data)
            # now sort the data in place
        data.sort(reverse=descending)
        for ix, item in enumerate(data):
            tree.move(item[1], '', ix)
        # switch the heading so it will sort in the opposite direction
        tree.heading(col, command=lambda col=col: sortby(tree, col, \
            int(not descending)))
```

```python
class Application:
    def __init__ (self):
        self.root = Tk()
        self.root.title("Sistema de Apoio a Saude Mental")
        self.root.option_add("*tearOff", FALSE)
        
        self.root.rowconfigure(0, weight=1)
        self.root.columnconfigure(0, weight=1)
        self.root.minsize(width=600, height=480)
        
        #Informação de sessão = (usuário, tipo, permissao)
        #Modificar mais quando inserir conexão com BD e criptografia
        self.session_info = (None, None, None)
        
        self.cur_frame = None
        
        self.generateMenuBar()
        self.generateAll()
        
        self.root.mainloop()
    def generateAll(self):
        
        self.Login = FormLogin(self.root)
        self.Login.LoginButton['command']=self.CommandLogin
        
        self.cur_frame = self.Login
        self.cur_frame.grid()
        
        self.Paciente = FormPaciente(self.root)
        self.Paciente.CancelButton['command']=self.CommandCancel
        
        self.Medico = FormMedico(self.root)
        self.Cronograma = Cronograma(self.root)
        
        self.BuscaPaciente = BuscaPaciente(self.root)
    def generateMenuBar(self):
        self.MenuBar = Menu(self.root)
        self.MenuPaciente = Menu(self.MenuBar)
        self.MenuMedico = Menu(self.MenuBar)
        self.MenuFuncionario = Menu(self.MenuBar)
        
        self.MenuBar.add_cascade(label="Paciente", menu=self.MenuPaciente)
        self.MenuBar.add_cascade(label="Medico", menu=self.MenuMedico)
        self.MenuBar.add_cascade(label="Funcionario", menu=self.MenuFuncionario)
        self.MenuBar.add_command(label="Logout", command=self.CommandLogout)
        
        self.MenuPaciente.add_command(label="Cadastrar Novo", command=self.CommandNovoPaciente)
        self.MenuPaciente.add_command(label="Buscar Paciente", command=self.CommandBuscarPaciente)
        self.MenuPaciente.add_command(label="Editar Paciente")
        self.MenuPaciente.add_command(label="Marcar consulta")
        self.MenuPaciente.add_command(label="Marcar exame")
        #self.MenuPaciente.add_command(label="Confirmar presença")
        #self.MenuPaciente.add_command(label="Ver histórico de consultas")
        #self.MenuPaciente.add_command(label="Ver histórico de exames")
        #self.MenuPaciente.add_command(label="Ver relatórios de consultas")
        #self.MenuPaciente.add_command(label="Ver resultados de exames")
        #self.MenuPaciente.add_command(label="Ver médicos associados")
        #self.MenuPaciente.add_command(label="Cadastrar receita")
        #self.MenuPaciente.add_command(label="Ver receitas")
        
        self.root['menu']=self.MenuBar
    def CommandLogin(self):
        user = self.Login.UserString.get()
        password = self.Login.PassString.get()
        
        self.Login.UserString.set('')
        self.Login.PassString.set('')
        
        if (user=="adm" and password=="adm"):
            self.sessionuser = (user, "adm", "adm")
            self.cur_frame.grid_remove()
            self.cur_frame = self.Cronograma
            self.cur_frame.grid()
    def CommandLogout(self):
        self.cur_frame.grid_remove()
        self.sessionuser = (None, None, None)
        self.cur_frame = self.Login
        self.cur_frame.grid()
    def CommandNovoPaciente(self):
        self.cur_frame.grid_remove()
        self.cur_frame = self.Paciente
        self.cur_frame.grid()
    def CommandCancel(self):
        self.cur_frame.grid_remove()
        self.cur_frame = self.Cronograma
        self.cur_frame.grid()
    def CommandBuscarPaciente(self):
        self.cur_frame.grid_remove()
        self.cur_frame = self.BuscaPaciente
        self.cur_frame.grid()

app = Application()
```
